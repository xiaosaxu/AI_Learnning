##### 初始化相关设置

```python
import requests as rq
import time
import warnings
warnings.filterwarnings("ignore")
from bs4 import BeautifulSoup
```

```python
headers_define = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
```

```python
def find_video(first_list,second_end_num):
    second_num = 0
    with open('video_info.txt','a') as f:
        for first_num in first_list:
            for second_num in range(second_end_num):
                july_url = "http://www.julyedu.com/video/play/"+str(first_num)+'/'+str(second_num)
                result2 =  rq.get(july_url,headers = headers_define)
                if result2.status_code == 200:
                    soup = BeautifulSoup(result2.content, fromEncoding="utf-8") #注意这个地方
                    samples = soup.find("li", "active")
                    print(samples)
                    if samples:
                        samples.get_text()
                        video_name = samples.get_text()
                        print(july_url+"   "+video_name)
                        f.write(july_url+"   "+video_name+'\n')
                        time.sleep(1)
                    else:
                        print(july_url+"   "+"NULL")
                        f.write(july_url+"   "+"NULL"+'\n')
                        time.sleep(1)
                else:
                    print(july_url +"return: "+str(result2.status_code))
                    f.write(july_url +"return: "+str(result2.status_code+'\n'))
                    time.sleep(1)
```

```python
find_video([38],1000)
```

```python
with open('video_info.txt','r') as f:
    for line in f:
        print (line.rstrip()) 
```
