## numpy 的综合应用 : 计算一个softmax矩阵

softmax 矩阵的意思是 每一个元素求其e的指数，在除以整行e的指数各个元素的和，得到（0-1）之间的数

随机生成10x10的矩阵

```python
m = np.random.rand(10,10)*10+4000
print(m)
```

```python
m_exp = np.exp(m)
m_exp
```

可以发现直接求e的n次方超出范围了，因此每行减去最大值再求。  
先求出每行的最大值

```python
m_row_max = m.max(axis=1).reshape(10,1)
m_row_max
```

使用broadcast特性，每行的每个元素减去每行的最大值

```python
m_midd = m - m_row_max
m_midd
```

求每个元素的e指数，使用np提供的函数 np.exp

```python
m_exp = np.exp(m_midd)
m_exp
```

求出每行的和，要注意后面的reshape

```python
m_row_sum = m_exp.sum(axis =1).reshape(10,1)
m_row_sum
```

同样使用broadcast特性，使用每行的元素除以每行的和

```python
m_result = m_exp/m_row_sum
m_result
```

下面验证一下结果，应该是每行的元素只和等于1

```python
m_result.sum(axis=1)
```

## python 中的列表切片

__列表的下标有三个参数：beg（起始下标），end（终止下标），delta（变化量）__
- 当delta未给出时：delta默认为1
- 当delta大于0时，beg默认为0，end默认为最末之后
- 当delta小于0时，beg默认为len（array）-1，end默认为开头之前

```python
lieb = [1,2,3,4,5,6,7]
```

delta 截取的间隔默认为一

```python
out = lieb[0:len(lieb):1]
out
```

delta 默认值为1

```python
out = lieb[::]
out
```

```python
out = lieb[:]
out
```

对于索引为负数的情况，会自动加上len() 转化为正数

```python
out = lieb[1:-1]
out
```

delta 为负数 表示截取的方向为从后往前间隔为delta的绝对值，此时需要保证开始的索引大于结束的索引

```python
out =  lieb[::-1]
out
```

```python
out = lieb[5:1:-1]
out
```

从上面例子可以看出来：即使是方向相反 python 的区间索引范围还是按照 __左开右闭__ 的形式

```python
out = lieb[::2]
out
```

如果方向不同，则返回空列表  
if (end - beg) * delta <= 0: return [] __
